   # Gender Split

Records the % female for each year where a count inlcudes a male and female breakdown.
(Note changes in Gender split occur when the number of one gender grows faster than the other.  It is possible to have growth in the number of female riders, but have the gender split lose a few percentage points if the number of male riders grows faster than the number of female riders.

```python
# Defaults
datadir = './script_output/count_observations/'

moves = ['north_turn_left','north_through','north_turn_right',\
         'east_turn_left','east_through', 'east_turn_right',\
         'south_turn_left','south_through','south_turn_right',\
         'west_turn_left','west_through','west_turn_right']

    # Moves from 'Single Site Data Analysis' work here.

goodcols = ['countsite', 'time', 'gender'] + moves
```

```python
import pandas as pd
import glob
import datetime as dt
from __future__ import division
```

```python
sites = datadir + '*'

sitelist = []
for f in glob.glob(sites):
    sitedir = f +'/*'
    datelist = []    
    for g in glob.glob(sitedir):
        obscsv = g + '/*'
        for h in glob.glob(obscsv):
            df = pd.read_csv(h, sep=', ', header = 0, usecols = goodcols, parse_dates=[0], \
                               infer_datetime_format=True, engine = 'python')
            site = df['countsite'][0]
            
            countdate = df['time'][0]
            countdate = dt.datetime.strptime(countdate,  "%Y-%m-%d %H:%M:%S")
            countdate = countdate.year
            
            females = df.query('gender == "F"')
            femaleresult = females.sum(axis =1, numeric_only=True)
            femaletotal = femaleresult.sum()
            
            males = df.query('gender == "M"')
            maleresult = males.sum(axis =1, numeric_only=True)
            maletotal = maleresult.sum()
            
            riders = femaletotal + maletotal
            # Convert ints to floats for division
             
            try: 
                pcfem = femaletotal / riders
                pcfem = round((pcfem*100),2)
            except:
                pcfem = "NaN"
                
            genderdf = pd.DataFrame(columns = [countdate])
            genderdf.loc[site] = [pcfem]
            datelist.append(genderdf)
            
        sitedf = pd.concat(datelist, axis = 1)
    sitelist.append(sitedf)
allsites = pd.concat(sitelist)

allsites   
```
